```python
import pandas as pd
import numpy as np
import datetime
```

```python
url = 'https://raw.githubusercontent.com/Agathamweb234/Alpha-Squad-3-Project/main/Train.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
np.shape(df)
```

```python
df.describe()
```

```python
df['STD'] = pd.to_datetime(df['STD'])
df.head()
```

```python
df['STA'] = pd.to_datetime(df['STA'], format = '%Y-%m-%d %H.%M.%S')
df.head()
```

```python
df['target'] = pd.to_datetime(df['target'])
df['target'] = pd.to_timedelta(df['target'].astype('int64'), unit= 'm')
df.head()
```

```python
df['Actual_Arrival'] = df['STA'] + df['target']
df.head()
```
